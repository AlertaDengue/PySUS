# Working with Infogripe datasets
[Infogripe](http://info.gripe.fiocruz.br/) is an online platform that tracks epidemiolgical data about influenza-like diseases in Brazil.


```python
from pysus.online_data.Infogripe import list_datasets, download
```

Infogripe makes available different datasets. To findout which ones are available before downloading we can ask PySUS to list them:

```python
list_datasets()
```

```python
%time df_alert = download("Alerta de situação")
df_alert
```

```python
df_age_sex_virus = download("Casos por idade, sexo e virus")
df_age_sex_virus
```

```python
df_totals = download("Casos Totais e estimativas")
df_totals
```

```python
df_localities = download("Valores esperados por localidades")
df_localities
```
