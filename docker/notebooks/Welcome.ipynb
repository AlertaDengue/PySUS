# Welcome to PySUS

PySUS provides tools for dealing with Brazil's public health data (SINAN, SINASC, SIM, SIH, SIA, PNI, IBGE, CNES, CIHA).

## Quick start

List available datasets and files:

```python
import pysus

pysus.list_files()
```

### Fetch SINAN data

```python
from pysus import sinan

df = sinan("deng", year=2023)
df.head()
```

### Fetch SINASC data

```python
from pysus import sinasc

df = sinasc(state="RJ", year=2023)
df.head()
```
